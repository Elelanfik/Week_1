 Checking and Handling Missing Values


```python
# List of datasets
datasets = [data_AAPL, data_AMZN, data_GOOG, data_MSFT, data_NVDA, data_TSLA, data_META]

# Corresponding dataset names (for output clarity)
dataset_names = ["AAPL", "AMZN", "GOOG", "MSFT", "NVDA", "TSLA", "META"]

# Loop through datasets with enumerate
for i, data in enumerate(datasets):
    print(f"Checking missing values for {dataset_names[i]}:")
    
    # Check for missing values
    missing_count = data.isnull().sum().sum()  # Total number of missing values
    if missing_count > 0:
        print(f" - Missing values found: {missing_count}")
        data.dropna(inplace=True)  # Drop rows with missing values
        print(f" - Missing values dropped.")
    else:
        print(f" - No missing values.")
    
    print(  "-" * 50  )
```

### Data Visualization

### Univariate Analysis
Univariate analysis examines individual columns to summarize their main characteristics

```python


# List of datasets and their names
datasets = [data_AAPL, data_AMZN, data_GOOG, data_MSFT, data_NVDA, data_TSLA, data_META]
dataset_names = ['AAPL', 'AMZN', 'GOOG', 'MSFT', 'NVDA', 'TSLA', 'META']

# Columns for univariate analysis
columns = ['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume', 'Dividends', 'Stock Splits']

# Univariate Analysis Loop for Each Dataset
for idx, data in enumerate(datasets):
    # Create a new figure for each dataset
    plt.figure(figsize=(16, 12))  # Adjust figure size for the plots

    # First row with 3 columns (first 3 columns)
    for i, column in enumerate(columns[:3]):  # First 3 columns (Open, High, Low)
        plt.subplot(2, 4, i + 1)  # 2 rows, 4 columns, first row, first 3 columns
        plot_univariate(data, column, title=f"{dataset_names[idx]}: {column}")

    # Second row with 4 columns (next 4 columns)
    for i, column in enumerate(columns[3:]):  # Next 4 columns (Close, Adj Close, Volume, Dividends, Stock Splits)
        plt.subplot(2, 4, i + 4)  # 2 rows, 4 columns, second row, last 4 columns
        plot_univariate(data, column, title=f"{dataset_names[idx]}: {column}")

    # Adjust layout for better spacing
    plt.tight_layout()
    plt.show()
```

## Bivariate Analysis
Bivariate analysis focuses on the relationships between two variables to identify correlations or patterns.


```python
# List of datasets and their names
datasets = [data_AAPL, data_AMZN, data_GOOG, data_MSFT, data_NVDA, data_TSLA, data_META]
dataset_names = ['AAPL', 'AMZN', 'GOOG', 'MSFT', 'NVDA', 'TSLA', 'META']

# Bivariate Analysis Loop for Each Dataset
for idx, data in enumerate(datasets):
    plt.figure(figsize=(16, 12))  # Adjust figure size for the plots

    # Pairs of columns to analyze bivariate relationships
    pairs = [
        ('Volume', 'Close'),        # Volume vs Close price
        ('Volume', 'Adj Close'),    # Volume vs Adjusted Close
        ('High', 'Low'),            # High vs Low (volatility)
        ('Volume', 'High'),         # Volume vs High price
        ('Volume', 'Low'),          # Volume vs Low price
        ('Volume', 'Stock Splits'), # Volume vs Stock Splits
        ('Close', 'Stock Splits'),  # Close price vs Stock Splits
        ('Adj Close', 'Stock Splits')  # Adjusted Close vs Stock Splits
    ]
    
    # Plotting bivariate analysis (scatter plots)
    for i, (x_column, y_column) in enumerate(pairs):
        # Determine subplot position in the 2x4 grid
        plt.subplot(2, 4, i + 1)  # 2 rows, 4 columns
        plot_bivariate(data, x_column, y_column, title=f"{dataset_names[idx]}: {x_column} vs {y_column}")
    
    # Adjust layout for better spacing
    plt.tight_layout()
    plt.show()
```

### Advanced Analytical Insights


```python
# List of datasets and their names
datasets = [data_AAPL, data_AMZN, data_GOOG, data_MSFT, data_NVDA, data_TSLA, data_META]
dataset_names = ['AAPL', 'AMZN', 'GOOG', 'MSFT', 'NVDA', 'TSLA', 'META']

# Window size for the moving average
window = 5

# Loop through each dataset, calculate the moving average, and plot the data
for idx, data in enumerate(datasets):
    # Add the Moving Average column for each dataset
    data['Moving Average'] = calculate_moving_average(data['Open'], window=window)
    
    # Create a figure with 2 subplots (side by side)
    fig, ax = plt.subplots(1, 2, figsize=(14, 6))  # 1 row, 2 columns for side by side plots
    
    # Plot the stock data with moving average on the first subplot
    plot_stock_data(data, window, ax[0])
    ax[0].set_title(f"{dataset_names[idx]}: Stock Data with Moving Average")
    
    # Optional: You can add another plot on the second axis (ax[1]) if needed
    # For example, a plot of Close price vs Moving Average
    ax[1].plot(data['Close'], label='Close Price', color='green')
    ax[1].plot(data['Moving Average'], label=f'Moving Average (window={window})', color='red', linestyle='--')
    ax[1].set_title(f"{dataset_names[idx]}: Close Price with Moving Average")
    ax[1].set_xlabel("Date")
    ax[1].set_ylabel("Price")
    ax[1].legend()

    # Adjust layout for better spacing
    plt.tight_layout()
    plt.show()
```

```python
sma = talib.SMA(data['Close'], timeperiod=5)
data['SMA'] = sma  # Add the SMA to the DataFrame

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(data.index, data['Close'], label='Close Price', color='blue', marker='o')
plt.plot(data.index, data['SMA'], label='SMA (3-period)', color='orange', marker='x')

# Add labels, title, and legend
plt.title('Closing Prices and 3-Period SMA')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend()
plt.grid()
plt.show()
```
